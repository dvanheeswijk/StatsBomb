```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import glob

with open('data/competitions.json') as f:
    competitions = json.load(f)
```

```python
# Women's World Cup ID is 72

wwc_id = 72
```

```python
with open('data/matches/'+str(wwc_id)+'/30.json') as f:
    matches = json.load(f)
```

```python
matches[0]
```

```python
df = pd.DataFrame(matches)

df.head(2)
```

```python
def dict_split(d, col):
    return d[col].apply(pd.Series)

cols = ['home_team', 'away_team', 'competition_stage']


for c in cols:
    a = dict_split(df, c)
    df = pd.concat([df, a], axis=1)
    df.drop(c, axis=1, inplace=True)

df.columns
```

```python
df.drop(['competition', 'season', 'match_status','metadata', 'last_updated'], axis=1, inplace=True)

df.head()
```

```python
df.drop(['stadium', 'referee'], axis=1, inplace=True)
df.columns
```

```python
names = ['match_id', 'match_date', 'kick_off', 'home_score', 'away_score', 'match_week', 'home_team_id',
        'home_team_name', 'home_team_gender', 'home_team_group', 'home_team_country',
        'home_team_managers', 'away_team_id', 'away_team_name', 'away_team_gender',
        'away_team_group', 'away_team_country', 'away_team_managers', 'competition_stage_id',
        'competition_stage_name']

df.columns = names

df.drop(['home_team_gender', 'away_team_gender', 'home_team_country', 
         'away_team_country', 'home_team_managers', 'away_team_managers'], axis=1, inplace=True)

df.head(2)
```

```python
df['home_team_name'].unique()
```

```python
df['home_team_name'] = df['home_team_name'].str.replace(" Women's", '')

df['home_team_name'].unique()
```

```python
df['away_team_name'] = df['away_team_name'].str.replace(" Women's", '')

df['away_team_name'].unique()
```
